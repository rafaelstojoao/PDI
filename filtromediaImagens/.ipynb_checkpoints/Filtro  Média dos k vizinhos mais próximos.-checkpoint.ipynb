#  Vamos modularizar isso...  fazer uma função que pode ser usada de qualquer lugar

```python
def qualMaisProximo(listaValores, chave):
#     print(chave,listaValores)
    maisProximo = 999999999.99
    distanciamaisProximo = 999999999.99
    posicaoMaisProximo = 9999

    for i in range(len(listaValores)):
        if( abs(listaValores[i]-chave) < distanciamaisProximo):
            maisProximo = listaValores[i]
            distanciamaisProximo = abs(listaValores[i]-chave)
            posicaoMaisProximo = i
            
    return maisProximo

 
    
numeros = np.random.rand(100) #100 valores float aleatórios


qualMaisProximo(numeros,0.43)
```

```python
# #considerando a vizinhança 3x3

# -----------------
# |  |  |  |  |  |
# |  |  |v0|v1|v2|
# |  |  |v3| x|v4|
# |  |  |v5|v6|v7|
# |  |  |  |  |  |
# -----------------

# Se x está em i,j, seus vizinhos são:
#     i-1,j-1
#     i-1,j
#     i-1,j+1
#     i,j-1
#     i,j+1
#     i,j-1
#     i+1,j
#     i+1,j+1

for i in range(len(im1)):
    for j in range(len(im1[i])):
        vizinhos =[]
        valorM1 = im1[i][j] #         valor do pixem em im1
        
        vizinhos.append(im1[max(i-1,0)][max(j-1,0)])
        vizinhos.append(im1[max(i-1,0)][j])
        vizinhos.append(im1[max(i-1,0)][min(j+1,278)])
        vizinhos.append(im1[i][max(j-1,0)])
        vizinhos.append(im1[i][min(j+1,278)])
        vizinhos.append(im1[min(i+1,357)][max(j-1,0)])
        vizinhos.append(im1[min(i+1,357)][j])
        vizinhos.append(im1[min(i+1,357)][min(j+1,278)])
        
        maisProximo = qualMaisProximo(vizinhos,valorM1)
        print("O valor mais proximo de ",valorM1,' dentro da vizinhança de 3x3 pixels é o: ', maisProximo)
        #é preciso pegar os k valores mais próximos de valorM1 numa janela de i+Janela, i-Janela
        im1[i][j] = maisProximo
        
        
    
plt.imshow(im1)
plt.show() 
```
