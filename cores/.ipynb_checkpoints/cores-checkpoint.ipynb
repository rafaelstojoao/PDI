```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import make_lupton_rgb

```

```python

r = 255/255
g = 255/255
b = 0/255

cor = [[r],[g],[b]]

image = make_lupton_rgb(cor[0],cor[1],cor[2], stretch=0.5)

plt.imshow(image)
```

```python

r = 255/255
g = 1/255
b = 200/255
cor1 = [[r],[g],[b]]


r = 0/255
g = 255/255
b = 0/255
cor2 = [[r],[g],[b]]

novaCor = []
for i in range(3):
    if cor1[i] == cor2[i] and cor1[i] == [1.0]:
        novoComponente = 1.0
    else:
        novoComponente = cor1[i][0]+cor2[i][0]
    novaCor.append([novoComponente])
    

cor = novaCor
image = make_lupton_rgb(cor[0],cor[1],cor[2], stretch=0.5)

plt.imshow(image)
```

# Com paciência, você pode construir imagens bitmap pela definição das cores

# Mas por que controlar as cores é importante?

```python
import cv2 as cv
import numpy as np

image = cv.imread("../fazendinha.png")
hsv=cv.cvtColor(image,cv.COLOR_BGR2HSV)

# Define lower and uppper limits of what we call "brown"
brown_lo=np.array([40,0,0])
brown_hi=np.array([120,255,255])


mask=cv.inRange(hsv,brown_lo,brown_hi)
image[mask>0]=(0,0,255)

cv.imwrite("resultfazenda.png",image)
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

while(1):
    _, tela = cap.read() #captura de imagem em frames

    # Conversão de RGB para HSV
    hsv = cv2.cvtColor(tela, cv2.COLOR_BGR2HSV)

    
    
    # Define um intervalo de tonalidades de azuis
    lower_blue = np.array([110,30,30])
    upper_blue = np.array([130,255,255])

    
    
    # Threshold the HSV image to get only blue colors
    mascara = cv2.inRange(hsv, lower_blue, upper_blue)
    #  frame_threshold = cv.inRange(frame_HSV, (low_H, low_S, low_V), (high_H, high_S, high_V))

    # Operador lógico AND aplicado a cada pixel (bit) por meio de uma máscara
    resultado = cv2.bitwise_and(tela,tela, mask= mascara)

    cv2.imshow('quadro',tela)
    cv2.imshow('mascara',mascara)
    cv2.imshow('Resultado',resultado)
    k = cv2.waitKey(5) & 0xFF
    
    
    if k == 27:
        break

cv2.destroyAllWindows()
```
